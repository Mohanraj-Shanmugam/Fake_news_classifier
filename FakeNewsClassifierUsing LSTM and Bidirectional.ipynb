```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
voc_size = 5000
```

```python
messages = X.copy()
```

```python
messages['title']
```

```python
messages.reset_index(inplace=True)
```

```python
messages['title']
```

```python
onehot_rep = [for words in df['text']]
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]',' ',messages['title'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(words) for words in review if words not in set(stopwords.words('english'))]
    corpus.append(' '.join(review))
    
corpus    
```

```python
onehot_repr = [one_hot(words, voc_size) for words in corpus]
onehot_repr
```

### Embedding Representation

```python
sent_length = 20
embedd_doc = pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
embedd_doc
```

```python
embedd_doc[0]
```

```python
from tensorflow.keras.models import Sequential
```

```python
embedding_vector_feature = 40
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_feature,input_length = sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.layers import Bidirectional
```

```python
embedding_vector_feature = 40
model1 = Sequential()
model1.add(Embedding(voc_size,embedding_vector_feature,input_length = sent_length))
model1.add(Bidirectional(LSTM(100)))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model1.summary()
```

```python
len(embedd_doc), y.shape
```

```python
import numpy as np
```

```python
X_final = np.array(embedd_doc)
y_final = np.array(y)
```

```python
X_final.shape, y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final,y_final,test_size=0.33,random_state=42)
```

Model_Training

```python
model1.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred1 = model1.predict_classes(X_test)
```

```python
y_pred = model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(confusion_matrix(y_test, y_pred1))
confusion_matrix(y_test, y_pred)
```

```python
print(accuracy_score(y_test, y_pred1))
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred1))
print(classification_report(y_test, y_pred))
```
